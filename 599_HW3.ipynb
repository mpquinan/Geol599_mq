```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.dates as mdates
```

```python
AQS = pd.read_csv('LA_AQS_2023.csv')
```

```python
AQS = AQS.rename(columns = {'Parameter Name':'parameter', 'Duration Description':'duration', 'Date (Local)':'date', 'Arithmetic Mean':'mean'})
```

```python
Oz_AQS = AQS.query('parameter == "Ozone" and duration == "1 HOUR"')
```

```python
Oz_AQS.head()
```

```python
fig, o = plt.subplots(figsize=(8, 6))
o = sns.scatterplot(x="date", y="mean", data=Oz_AQS,)
o.set(xlabel='Date', ylabel='Nitrogen Dioxide (ppb)')
month_locator = mdates.MonthLocator(interval=1) 
o.xaxis.set_major_locator(month_locator)
fig.autofmt_xdate()
```

```python
Oz35_AQS = Oz_AQS.query('mean > 0.035')
```
